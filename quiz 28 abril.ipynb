```python
import csv

datos = [
    ["nombre", "edad", "pais"],
    ["Marieth", 20, "colombia"],
    ["Samuel", 23, "chile"],
    ["David", 22, "argentina"]
]

with open("datos.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(datos)
```
