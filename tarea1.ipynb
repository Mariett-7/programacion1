# (1). *Listas a diccionario, tupla y conjunto*

```python
lista_universidades = ["UIS", "UCC", "UTS", "UDI", "UNAB", "USTA"]
print(lista_universidades)
```

```python
universidades = ["UIS", "UCC", "UTS", "UDI", "UNAB", "USTA"]
tupla = tuple(universidades)
diccionario = dict(enumerate(universidades))
conjunto = set(universidades)
print(tupla,diccionario,conjunto)
```

# (2).*Tupla a lista, diccinario y conjunto*

```python
tupla_materias = ("análisis", "moderna", "estadística", "programación")
print(tupla_materias)
```

```python
materias = ("análisis", "moderna", "estadística", "programación")
lista = list(materias)
diccionario = dict(enumerate(materias))
conjunto = set(materias)
print(lista, diccionario, conjunto)
```

#(3). *Diccionario a lista, tupla y conjunto*

```python
diccionario_edificios = {"CT": "Camilo Torrres", "CH": "Ciencias Humanas", "II": "Ingenieria Industrial"}
print(diccionario_edificios)
```

```python
edificios = {"CT": "Camilo Torrres", "CH": "Ciencias Humanas", "II": "Ingenieria Industrial"}

# con las claves de mi diccionario
lista = list(edificios)
tupla = tuple(edificios)
conjunto = set(edificios)
print(lista, tupla, conjunto)

# con los valores de mi diccionario
edificios.values()

lista1 = list(edificios.values())
tupla1 = tuple(edificios.values())
conjunto1 = set(edificios.values())
print(lista1)
print(tupla1)
print(conjunto1)

# con las tuplas (clave, valor) de mi diccionario
edificios.items()

lista2 = list(edificios.items())
tupla2 = tuple(edificios.items())
conjunto2 = set(edificios.items())
print(lista2)
print(tupla2)
print(conjunto2)
```
