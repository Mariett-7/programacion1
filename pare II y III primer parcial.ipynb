```python
A= {2,3,4,5}
B = {2,3,7,6}
A.difference(B)
```

la operacion realizada por la sentencia *difference* entre 2 conjuntos toma los elementos de mi conjunto inicial(A), sin contar los elementos que estan en el otro conjunto(B). Así el conjunto resultante es los elementos que estan en A, pero no estan en B, eso es A\B= {4,5]

```python
A.symmetric_difference(B)
```

la operación realizada por la sentencia *symmetric_difference* entre 2 conjuntos lo que hace, es que toma los valores del conjunto A y B sin tomar los que están en la intersección de estos mismos. para el caso el conjunto formado al sacar la diferencia simetrica entre A y B es {4,5,6,7}, es decir, A unido B, menos A intersección B.
