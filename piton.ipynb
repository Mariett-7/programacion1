CONTROL DE FLUJO EN PYTHON

```python
# series de fibonacci
#la suma de dos elementos define el siguiente
a=0
b=1
while b<100: # while establece el limite de hasta donde controlaremos el flujo
  print(b)
  a,b = b, a+b
```

```python
a=0
b=1
while b<100: # while establece el limite de hasta donde controlaremos el flujo
  print(b)
  a=b
  b=a+b
```
