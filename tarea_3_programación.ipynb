```python
def sumar_vectores(A, B):
    n = len(A)
    if n != len(B):
        return "Error: Los vectores deben tener las mismas dimensiones para sumarse."
    else:
        try:
            resultado = []
            for i in range(n):
                resultado.append(float(A[i]) + float(B[i]))
            return resultado
        except ValueError:
            return "Error: Debe ingresar solo numeros."


def producto_punto(A, B):
    n = len(A)
    if n != len(B):
        return "Error: Los vectores deben tener las mismas dimensiones para el producto punto."
    else:
        try:
            resultado = 0
            for i in range(n):
                resultado += float(A[i]) * float(B[i])
            return resultado
        except ValueError:
            return "Error: Debe ingresar solo numeros."


def producto_cruz(A, B):
    if len(A) != 3 or len(B) != 3:
        return "Error: El producto cruz solo esta definido para vectores de 3 dimensiones."
    else:
        try:
            a0, a1, a2 = float(A[0]), float(A[1]), float(A[2])
            b0, b1, b2 = float(B[0]), float(B[1]), float(B[2])
            resultado =[
                a1 * b2 - a2 * b1,
                a2 * b0 - a0 * b2,
                a0 * b1 - a1 * b0]
            return resultado
        except ValueError:
            return "Error: Debe ingresar solo numeros."


def hadamard(A, B):
    n = len(A)
    if n != len(B):
        return "Error: Los vectores deben tener las mismas dimensiones para el producto Hadamard."
    else:
        try:
            resultado = []
            for i in range(n):
                resultado.append(float(A[i]) * float(B[i]))
            return resultado
        except ValueError:
            return "Error: Debe ingresar solo numeros."


while True:
    try:
        print("      ---OPERACIONES ENTRE VECTORES---     ")
        print("---------- ")
        print("1. Suma de vectores")
        print("2. Producto punto")
        print("3. Producto cruz")
        print("4. Producto de Hadamard")
        print("5. Salir")
        print("---------- ")


        opcion = int(input("Elige una opcion: "))

        if opcion == 1:
            v1 = input("Ingrese el vector 1 separado por espacios: ").split()
            v2 = input("Ingrese el vector 2 separado por espacios: ").split()
            r = sumar_vectores(v1, v2)
            print("La suma es:", r)

        elif opcion == 2:
            v1 = input("Ingrese el vector 1 separado por espacios: ").split()
            v2 = input("Ingrese el vector 2 separado por espacios: ").split()
            r = producto_punto(v1, v2)
            print("El producto punto es:", r)

        elif opcion == 3:
            v1 = input("Ingrese el vector 1 separado por espacios: ").split()
            v2 = input("Ingrese el vector 2 separado por espacios: ").split()
            r = producto_cruz(v1, v2)
            print("El producto cruz es:", r)

        elif opcion == 4:
            v1 = input("Ingrese el vector 1 separado por espacios: ").split()
            v2 = input("Ingrese el vector 2 separado por espacios: ").split()
            r = hadamard(v1, v2)
            print("El producto Hadamard es:", r)

        elif opcion == 5:
            print("Fin del programa")
            break

        else:
            print("Opcion incorrecta, ingrese un numero entre 1 y 5.")

    except ValueError:
        print("Opcion incorrecta, debe ingresar un numero entero.")
```
